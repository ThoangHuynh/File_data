##**K-Means**
The k-means clustering algorithm is a simple but effective algorithm for cluster stock labeling. k-means clustering works by dividing the data into k clusters, where k is a user-defined parameter. The algorithm starts by randomly assigning each data point to one of the k clusters. Then, the algorithm iteratively updates the cluster centroids until the algorithm converges.

The Elbow Method is a popular technique for finding the optimal number of clusters. It works by plotting the distortion of the clusters as a function of the number of clusters. The optimal number of clusters is the point at which the distortion starts to decrease more slowly.

```python
from sklearn.cluster import KMeans

# Create an empty list to store the SSE values
sse = []

# Define the range of k values to try
K = range(1, 10)

# Iterate over the range of k values
for k in K:
    # Create a KMeans instance with k clusters
    kmeanModel = KMeans(n_clusters=k)
    
    # Fit the KMeans instance to the standardized data
    kmeanModel.fit(dt_pre_std)
    
    # Append the SSE (inertia) value to the sse list
    sse.append(kmeanModel.inertia_)

# Create a plot of k vs. SSE
plt.figure(figsize=(9, 9))
plt.plot(K, sse, marker='o', color='r')
plt.xticks(K)
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

#**PCA**
Principal component analysis, or PCA, is a statistical procedure that allows you to summarize the information content in large data tables by means of a smaller set of “summary indices” that can be more easily visualized and analyzed

The plot will show the cumulative explained variance as more principal components are included. We can see how many components are needed to explain 95% of the variance.

```python
from sklearn.decomposition import PCA

# Standardize the data and fit a PCA model
pca = PCA().fit(dt_pre_std)

# Compute the cumulative explained variance ratio and plot it
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))

# Define the x and y axis labels and the plot title
plt.xlabel('Number of Components', fontsize=15)
plt.ylabel('Variance (%)', fontsize=15) 
plt.title('Explained Variance', fontsize=20)

# Show the plot
plt.show()
```

There are 2 number of components

```python
# Perform PCA with two principal components
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(dt_pre_std)

# Create a new DataFrame with the principal components
pca = pd.DataFrame(data = principalComponents, columns = ['PCA1','PCA2'], index=dt_pre.index)
```

Visualize the principal components.
You can visualize the principal components on a 2D scatter plot:

```python
# Create a new figure
plt.figure()

# Set the x and y axis labels and plot title
plt.xlabel('PCA1', fontsize=15)
plt.ylabel('PCA2', fontsize=15)
plt.title('Component PCA', fontsize=20)

# Annotate each point with its label
for i in range(len(pca)):
    x, y = pca.iloc[i, 0], pca.iloc[i, 1]
    label = pca.index[i]
    plt.text(x, y, label, color='black', fontsize=10)

# Plot the scatter plot of the principal components
plt.scatter(pca['PCA1'], pca['PCA2'], c='royalBlue', s=50)
```

By running PCA, we have reduced the number of dimensions in our data set from 3 to 2; this means if we graph our new data frame, we will only have two dimensions. Let's see how the new data set would look graphically.

CATEGORIES VISUALIZATION

```python
# Perform k-means clustering with 2 clusters on the PCA transformed data
kmeans = KMeans(n_clusters=2, random_state=0).fit(pca)

# Get the cluster centers
cluster_centers = kmeans.cluster_centers_
center1 = cluster_centers[:, 0]
center2 = cluster_centers[:, 1]

# Create a new 2D plot
plt.figure()

# Set the axes labels
plt.xlabel(pca.columns[0])
plt.ylabel(pca.columns[1])

# Plot the data points with different colors based on the cluster assignment
plt.scatter(pca['PCA1'], pca['PCA2'], c=kmeans.labels_.astype(float), cmap='Accent')

# Plot the cluster centers as blue crosses
plt.scatter(center1, center2, marker="x", color='b')

# Annotate each point with its label
for i in range(len(pca)):
    x, y = pca.iloc[i, 0], pca.iloc[i, 1]
    label = pca.index[i]
    plt.text(x, y, label, color='black', fontsize=10)

# Set the title of the plot
plt.title('Visualization of clustered data with {} clusters'.format(2), fontweight='bold')

# Show the plot
plt.show()
```

The plot shows that the data is well-clustered, with two distinct clusters. The cluster centers are located in the center of each cluster. This suggests that the PCA transformation was successful in identifying the two main underlying factors that drive the variation in the data.

The two clusters can be interpreted as follows:

* Cluster 1: This cluster contains points that are high on PCA1 and low on PCA2. These points may be characterized by high stock prices and low volatility.
* Cluster 2: This cluster contains points that are low on PCA1 and high on PCA2. These points may be characterized by low stock prices and high volatility.

```python
# Define a function to extract closing prices from a list of dataframes
def closing_price(a):
    # Create a series to hold the closing prices
    closing_price = pd.Series(data[0].iloc[-1, 29+a])
    # Iterate over each dataframe in the list and append the closing price to the series
    for i in range(1, len(data)):
        temp = data[i]
        closed_price = temp.iloc[-1, 29+a]
        series_to_concat = pd.Series(closed_price)
        closing_price = pd.concat([closing_price, series_to_concat])
    # Return the series of closing prices
    return closing_price
```
